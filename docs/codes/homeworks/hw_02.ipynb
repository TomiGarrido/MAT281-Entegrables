```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

product_df = pd.read_csv("products_info.csv")
district_df = pd.read_csv("districts_info.csv")
display(product_df)
display(district_df)

most_common_state = district_df['state'].mode()[0]
most_common_locale = district_df['locale'].mode()[0]

district_df['state'].fillna(most_common_state, inplace=True)
product_df['Provider/Company Name'].fillna('Unknown', inplace = True)
product_df['Sector(s)'].fillna(product_df['Sector(s)'].mode()[0], inplace=True)
product_df['Primary Essential Function'].fillna(product_df['Primary Essential Function'].mode()[0], inplace=True)
```

```python
plt.figure(figsize=(18,10))

ax=sns.countplot(data=district_df,x='state',palette='pastel', hue='locale')

plt.xticks(rotation=45)
plt.title("Estado y sus tipos de localidad")
plt.legend(loc='upper right')
for p in ax.patches:
    ax.text (p.get_x() + p.get_width()  / 2,p.get_height()+ 0.25,p.get_height(), fontsize = 11)
```

```python
plt.figure(figsize=(16, 10))

sns.countplot(y="state",data=district_df,order=district_df.state.value_counts().index,palette="pastel",linewidth=3)
plt.title("Distribución de los estados en Estados Unidos",size=18)
sns.despine()
plt.show()

locale_counts = district_df['locale'].value_counts()
labels = locale_counts.index
sizes = locale_counts.values

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Distribución de tipos de Localidad en los distritos')
plt.axis('equal')
plt.show()
```

```python
# Simplemente vamos a procesar los datos de la carpeta "engagement_data"
temp = []
for district in district_df.district_id.unique():
    df = pd.read_csv(f'{"engagement_data"}/{district}.csv', index_col=None, header=0)
    df["district_id"] = district
    temp.append(df)

engagement = pd.concat(temp)
engagement = engagement.reset_index(drop=True)
engagement
```

```python
# Convertimos a fechas las columnas de "tiempo" y además rellenamos los valores vacios en "engagement_index"
engagement['time'] = pd.to_datetime(engagement['time'])
engagement['engagement_index'] = engagement['engagement_index'].fillna(0)

print("Columna 'time' convertida a datetime y valores NaN en 'engagement_index' rellenados con 0.")
display(engagement.head())

# Ahora vamos a fusionar el dataFrame 'engagement' con 'district_df' usando 'district_id' y con 'product_df'
merged_df_districts = pd.merge(engagement, district_df, on='district_id', how='left')
final_merged_df = pd.merge(merged_df_districts, product_df, left_on='lp_id', right_on='LP ID', how='left')

# Aquí vamos a ver cuáles fueron los productos mas usados durante 2020
temp_df = final_merged_df[['time', 'Product Name', 'engagement_index']].copy()
engagement_2020 = temp_df[temp_df['time'].dt.year == 2020]
if engagement_2020['engagement_index'].dtype == 'float64':
    engagement_2020['engagement_index'] = engagement_2020['engagement_index'].astype('float32')
avg_daily_engagement_product = engagement_2020.groupby('Product Name')['engagement_index'].mean().reset_index()
avg_daily_engagement_product_sorted = avg_daily_engagement_product.sort_values(by='engagement_index', ascending=False)

print("Top 5 productos mas utilizados en 2020")
display(avg_daily_engagement_product_sorted.head(5))
```

```python
# Aquí vamos a graficar las tendencias de "engagement_index" a lo largo del tiempo.
# Se considera cada día del año, desde el primer dia de 2020 hasta el primer día de 2021
avg_engagement_over_time = final_merged_df.groupby('time')['engagement_index'].mean().reset_index()
plt.figure(figsize=(15, 7))
sns.lineplot(x='time', y='engagement_index', data=avg_engagement_over_time)
plt.title('Tendencia del Índice de Engagement Promedio a lo Largo del Tiempo', fontsize=16)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Índice de Engagement Promedio', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression

# Aqui simplemente agregamos los archivos y se leen
train_data = pd.read_csv("train.csv")
display(train_data.head())

test_data = pd.read_csv("test.csv")
display(test_data.head())

train_data['Cabin'] = train_data['Cabin'].fillna('Unknown')
most_frequent_embarked = train_data['Embarked'].mode()[0]
train_data['Embarked'] = train_data['Embarked'].fillna(most_frequent_embarked)
imputer = SimpleImputer(strategy='median')
train_data['Age'] = imputer.fit_transform(train_data[['Age']])

features = ["Pclass", "Sex", "SibSp", "Parch"]

X_train = train_data[features].copy()
Y_train = train_data["Survived"]

X_test = test_data[features].copy()

X_train['Sex'] = X_train['Sex'].map({'male': 0, 'female': 1})
X_test['Sex'] = X_test['Sex'].map({'male': 0, 'female': 1})

display(X_train.head())
display(X_test.head())

# Aqui vamos enseñandole al modelo RandomForest cómo hacer la prediccion y guarda un archivo
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train, Y_train)
predictions_rf = model.predict(X_test)
submission = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': predictions_rf})
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions_rf})
output.to_csv('sobrevivientes_RF.csv', index=False)

# Es lo mismo que antes, pero lo ahcemos con Regresion Logistica y no con Random Forest
# Es como alternativa
model_lr = LogisticRegression(solver='liblinear', random_state=1)
model_lr.fit(X_train, Y_train)
predictions_lr = model_lr.predict(X_test)
submission1 = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': predictions_lr})
output1 = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions_lr})
output1.to_csv('sobrevivientes_LR.csv', index=False)
```
