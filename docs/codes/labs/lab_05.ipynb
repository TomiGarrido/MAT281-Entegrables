   * Muestra c칩mo se distribuyen los t칤tulos seg칰n su `release_year`.
   * Pregunta gu칤a: 쮿ay m치s contenido reciente o antiguo? 쯅otas alg칰n patr칩n en d칠cadas espec칤ficas?



## 游꿛 Parte 2: Tendencias y evoluci칩n en el tiempo

4. **N칰mero de t칤tulos agregados por a침o**

   * Usa `date_added` (convertido a fecha) y grafica una serie temporal.
   * Pregunta gu칤a: 쮺u치ndo creci칩 m치s el cat치logo de Netflix? 쯈u칠 relaci칩n podr칤a tener con la expansi칩n internacional de la plataforma?

5. **Heatmap de lanzamientos por a침o y mes**

   * Construye un mapa de calor con `year_added` y `month_added`.
   * Pregunta gu칤a: 쮿ay meses con m치s estrenos que otros? 쯈u칠 conclusiones puedes sacar sobre la estacionalidad de lanzamientos?

6. **Duraci칩n de pel칤culas por g칠nero**

   * Extrae la duraci칩n en minutos y crea un boxplot agrupado por g칠nero principal.
   * Pregunta gu칤a: 쯈u칠 g칠neros suelen durar m치s? 쯈u칠 podr칤as decir sobre la variabilidad dentro de un mismo g칠nero?


```python
"4."
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
df["year_added"] = df["date_added"].dt.year
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x="year_added", color="darkred")
plt.title("N칰mero de t칤tulos agregados por a침o")
plt.xlabel("A침o")
plt.ylabel("Cantidad de t칤tulos")
plt.show()
print("El cat치logo de Netflix creci칩 y lleg칩 a un m치ximo en el a침o 2019.")

"5."
df["month_added"] = df["date_added"].dt.month
cuenta_mensual = df.groupby(["year_added", "month_added"]).size().unstack(fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(cuenta_mensual, color="darkred", annot=True, fmt="d")
plt.title("Mapa de calor de t칤tulos por a침o y mes")
plt.xlabel("Mes")
plt.ylabel("A침o")
plt.show()
print("Si, hay meses con mas estrenos que otros.")
print("Sobre la estacionalidad se puede decir que las pel칤culas suelen estrenarse mas ")

"6."
movies_df = df[df["type"] == "Movie"].copy()
movies_df["duration_minutes"] = movies_df["duration"].str.replace(" min", "", regex=False).astype(float)
genres_df = movies_df.assign(listed_in=movies_df["listed_in"].str.split(", ")).explode("listed_in").reset_index(drop=True)
plt.figure(figsize=(14, 8))
sns.boxplot(data=genres_df, x="duration_minutes", y="listed_in", color="darkred")
plt.title("Distribuci칩n de duraci칩n de pel칤culas por g칠nero")
plt.xlabel("Duraci칩n (min)")
plt.ylabel("G칠nero")
plt.show()
print("Los g칠neros cl치sicos y musicales son los mas duraderos.")
```



## 游댠 Parte 3: Comparaciones y relaciones

   * Usa un gr치fico de barras para mostrar los pa칤ses m치s frecuentes en la columna `country`.
   * Pregunta gu칤a: 쯈u칠 pa칤ses dominan el cat치logo? 쯉orprende la distribuci칩n o es esperada?

8. **Pel칤culas vs Series seg칰n g칠nero**

   * Realiza un gr치fico de barras apiladas para mostrar la cantidad de t칤tulos por g칠nero, separando `Movie` y `TV Show`.
   * Pregunta gu칤a: 쮼xisten g칠neros m치s asociados a pel칤culas o a series?

9. **Relaci칩n entre duraci칩n y a침o de lanzamiento**

   * Haz un gr치fico de dispersi칩n (scatterplot) con `release_year` en el eje X y `duration` (en minutos) en el eje Y.
   * Pregunta gu칤a: 쮿an cambiado los est치ndares de duraci칩n de las pel칤culas con el tiempo? 쯉e ven outliers interesantes?



```python
"7."
country_counts = df["country"].value_counts().nlargest(10)
sns.barplot(x=country_counts.index, y=country_counts.values, color="darkred", legend=False)
plt.title("Top 10 Pa칤ses con m치s producciones en Netflix")
plt.xlabel("Pa칤s")
plt.ylabel("N칰mero de Producciones")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
print("La mayor cantidad de peliculas son producidas en Estados Unidos")
print("No sorprende, debido a que es justamente en estados unidos en donde se encuentran los mayores productores de peliculas.")

"8."
genres_df = df.assign(listed_in=df["listed_in"].str.split(", ")).explode("listed_in")
genre_type_counts = genres_df.groupby("listed_in")["type"].value_counts().unstack().fillna(0)
genre_type_counts.plot(kind="bar", stacked=True, figsize=(12, 7), color="darkred")
plt.title("Pel칤culas vs Series por g칠nero")
plt.xlabel("G칠nero")
plt.ylabel("N칰mero de t칤tulos")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
print("De acuerdo al gr치fico, existen ma치s g칠neros en pel칤culas que teleseries.")

"9."
movies_df = df[df["type"] == "Movie"].copy()
movies_df["duration_minutes"] = movies_df["duration"].str.replace(" min", "").astype(float)
plt.figure(figsize=(12, 6))
sns.scatterplot(data=movies_df, x="release_year", y="duration_minutes", alpha=0.6)
plt.title("Relaci칩n entre la duraci칩n de pel칤culas y a침o de lanzamiento")
plt.xlabel("A침o de lanzamiento")
plt.ylabel("Duracion (min)")
plt.tight_layout()
plt.show()
print("S칤,hay un cambio en los est치ndares sobre los est치ndares de la duraci칩n de pel칤culas.")
print("En efecto, se observan outliers interesantes, como una pel칤cula cuya duraci칩n es de +300 minutos.")
```

## 救 Desaf칤o Final

* Encuentra las combinaciones m치s frecuentes de **g칠nero + rating** y repres칠ntalas en un **heatmap o gr치fico de burbujas**.
* Pregunta gu칤a: 쯈u칠 g칠neros est치n m치s dirigidos a un p칰blico adulto? 쯏 cu치les a p칰blico familiar o infantil?

```python
genres_and_ratings = df.assign(listed_in=df["listed_in"].str.split(", ")).explode("listed_in")
genre_rating_counts = genres_and_ratings.groupby(["listed_in", "rating"]).size().unstack(fill_value=0)
top_genres = genres_and_ratings["listed_in"].value_counts().nlargest(10).index
genre_rating_counts_top = genre_rating_counts.loc[top_genres]
plt.figure(figsize=(12, 8))
sns.heatmap(genre_rating_counts_top, annot=True, fmt="d", color="darkred")
plt.title("Combinaciones m치s frecuentes de g칠nero y Clasificaci칩n por Edad")
plt.xlabel("Clasificaci칩n edad")
plt.ylabel("G칠nero")
plt.tight_layout()
plt.show()
print("Los g칠neros que est치n m치s dirigidos hacia el p칰blico adulto son de drama, comedias e internatcionales.")
print("Mientras que las que son dirigidas hacia el p칰blico familiar o infantil son tambi칠n dramas e internacionales.")
```
