```python
"7."
country_counts = df["country"].value_counts().nlargest(10)
sns.barplot(x=country_counts.index, y=country_counts.values, color="darkred", legend=False)
plt.title('Top 10 Pa√≠ses con M√°s Producciones en Netflix')
plt.xlabel('Pa√≠s')
plt.ylabel('N√∫mero de Producciones')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
print("La mayor cantidad de peliculas son producidas en Estados Unidos")
print("No sorprende, debido a que es justamente en estados unidos en donde se encuentran los mayores productores de peliculas.")

"8."
genres_df = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')

# Count the occurrences of each genre for movies and TV shows
genre_type_counts = genres_df.groupby('listed_in')['type'].value_counts().unstack().fillna(0)

# Plotting the stacked bar chart
genre_type_counts.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Pel√≠culas vs Series por G√©nero')
plt.xlabel('G√©nero')
plt.ylabel('N√∫mero de T√≠tulos')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# 9. Relaci√≥n entre duraci√≥n y a√±o de lanzamiento
# Filter for movies and extract duration in minutes
movies_df = df[df['type'] == 'Movie'].copy()
movies_df['duration_minutes'] = movies_df['duration'].str.replace(' min', '').astype(float)

# Create scatterplot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=movies_df, x='release_year', y='duration_minutes', alpha=0.6)
plt.title('Relaci√≥n entre Duraci√≥n de Pel√≠culas y A√±o de Lanzamiento')
plt.xlabel('A√±o de Lanzamiento')
plt.ylabel('Duraci√≥n (minutos)')
plt.tight_layout()
plt.show()
```
